```python
from IPython import get_ipython
from IPython.display import display
# %%
!pip install scikit-learn numpy matplotlib
# %%
import numpy as np  #for mathematical calculation like coordinates
import matplotlib.pyplot as plt #data visualization
from sklearn.datasets import fetch_openml #fetching data from dataset
from sklearn.manifold import TSNE #

#Load the MNIST dataset
mnist = fetch_openml('mnist_784', as_frame=False) # Added as_frame=False
X, y = mnist['data'], mnist['target']
X = X / 255.0 #Normalize pixel values to[0,1]

#Reduce the dimentionality using t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

#create the scatterplot to visualize the reduced data
plt.figure(figsize=(10, 8))
# Corrected colorbar and ticks
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y.astype(int), cmap='jet', s=10)
cbar = plt.colorbar(label='Digit')
cbar.set_ticks(np.arange(10))
cbar.set_ticklabels(np.arange(10))
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE visualization of MNIST dataset')
plt.show()
```
